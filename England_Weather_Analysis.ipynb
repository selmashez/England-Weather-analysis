```python
# Calculating the probability of temperatures above 20 degrees Celsius

hot_days = df[df['Temperature (C)'] > 20]

print(len(hot_days))

rows = df.shape[0]

print(rows)

p = len(hot_days) / rows

print("probibility of being a hot day in england :" , p*100)
```

This calculation shows that the probability of the temperature in London being above 20 degrees is 21 percent

```python
# Calculating the probability of the temperature is less than 10 degrees and the wind speed is less than 12

p_intersection = 100 * (len(df[(df['Temperature (C)'] < 10) & (df['Wind Speed (km/h)'] < 12)]) / len(df))

print(f"Probability of the temperature being less than 10 degrees and the wind speed being less than 12: {p_intersection:.2f}%")
```

This calculation examines the probability of the occurrence of temperatures below 10 degrees and wind speeds below 12 kilometers per hour together, which is the intersection probability of the two events.

```python
# Bayes' Theorem
#Calculate the probability that a pressure is more than 1010 given they have a Humidity less than 0.6

p_pressure_and_humadity = 100 * (len(df[(df['Pressure (millibars)'] > 1010) & (df['Humidity'] < 0.6)]) / len(df))

p_humadity_less = len(df[df['Humidity'] < 0.6 ]) / len(df)

p_bayes = p_pressure_and_humadity / p_humadity_less

print(f"the probability that a pressure is more than 1010 given they have a Humidity less than 0.6: {p_bayes:.2f}%")
```

In this calculation, using Bayes' theorem, I calculated the probability of the pressure being above 1010 while the humidity is below 0.6, which indicates that higher pressure will result in lower humidity.

```python
# Check Normality
# using histogram to check if Humadity is normally distributed

sns.histplot(df['Humidity'], kde=True, bins = 50)
plt.title('Histogram of Humidity Before Noise Removal')
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.show()
```

Here, I have examined the distribution of humidity data using a histogram, which shows that the data are not normally distributed

```python
# Using Shapiro-Wilk test to check if Humadity is normally distributed

statistic, p_value = stats.shapiro(df['Humidity'])
print(f"Shapiro-Wilk Test for Humidity: Statistic = {statistic}, p-value = {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Humidity is not normally distributed.")
else:
    print("Humidity is normally distributed.")
```

To further ensure accuracy, I used the Shapiro-Wilk test to check the normality of the data distribution. In this test, the closer the p-value is to zero, the more likely the assumption of normality is invalidated. Since the test result is zero, it clearly confirms the non-normality of the humidity data.
